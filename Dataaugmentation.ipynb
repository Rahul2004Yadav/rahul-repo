```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image

# Import ImageDataGenerator directly from tensorflow.keras.preprocessing.image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img("catdata.jpg",target_size=(200,200))
```

```python
img
```

```python
datagen=ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2
)
```

```python
img=image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
import os

# Create the directory if it doesn't exist
if not os.path.exists('catdata'):
    os.makedirs('catdata')

i=0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='catdata'):
  i=i+1
  if i == 10:
    break
```
